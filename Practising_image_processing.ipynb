**Resizing the Images using PIL**

```python
#load the images
img1 = Image.open('img1.jpg')
img2 = Image.open('img2.jpg')
img3 = Image.open('img3.jpg')
img4 = Image.open('img4.jpg')
```

```python
img1_resized = img1.resize((240,360))
img2_resized = img2.resize((240,360))
img3_resized = img3.resize((240,360))
img4_resized = img4.resize((240,360))
```

**Size of the resized images**


```python
img1_resized.size
img2_resized.size
img3_resized.size
img4_resized.size
```

**Showing resized images**

**Save the resized images**

```python
img1_resized.save('img1_resized.jpg')
img2_resized.save('img2_resized.jpg')
img3_resized.save('img3_resized.jpg')
img4_resized.save('img4_resized.jpg')
```

**Converting the resized images into grayscale**

```python
import cv2
```

Calling the resized images

```python
img1_resized = cv2.imread('img1_resized.jpg')
img2_resized = cv2.imread('img2_resized.jpg')
img3_resized = cv2.imread('img3_resized.jpg')
img4_resized = cv2.imread('img4_resized.jpg')
```

**Conversion**

```python
img1_resized_gray = cv2.cvtColor(img1_resized, cv2.COLOR_RGB2GRAY)
img2_resized_gray = cv2.cvtColor(img2_resized, cv2.COLOR_RGB2GRAY)
img3_resized_gray = cv2.cvtColor(img3_resized, cv2.COLOR_RGB2GRAY)
img4_resized_gray = cv2.cvtColor(img4_resized, cv2.COLOR_RGB2GRAY)
```

plotting the images

```python
from google.colab.patches import cv2_imshow
cv2_imshow(img1_resized_gray)
cv2_imshow(img2_resized_gray)
cv2_imshow(img3_resized_gray)
cv2_imshow(img4_resized_gray)
```

Saving the grayscale images

```python
cv2.imwrite('gray_resized_img1.jpg',img1_resized_gray)
cv2.imwrite('gray_resized_img2.jpg',img2_resized_gray)
cv2.imwrite('gray_resized_img3.jpg',img3_resized_gray)
cv2.imwrite('gray_resized_img4.jpg',img4_resized_gray)
```
